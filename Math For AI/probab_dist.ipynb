```python
import numpy as np
import matplotlib.pyplot as plt

#values btw 0 to 10
values=np.random.uniform(0,10,10_00_000)

#plotting a histogram- becomes more uniform with more values
plt.hist(values,bins=100,alpha=0.7,density=True)
plt.title("Continous Uniform Distribution between 0-10 ")
plt.xlabel("x")
plt.ylabel("Probability density")
plt.grid()
plt.show()
```

```python
# Normal Distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#parameters
mu=70 #mean
sigma=10 #standard deviation

#x-axis values
x=np.linspace(30,110,1000)

#Normal PDF
y=norm.pdf(x,mu,sigma)

#plot the bell curve
plt.plot(x,y,color='black',linewidth=2,label='Normal Distribution')

#Shade 68% for region (mu+-1)
x1=np.linspace(mu-sigma,mu+sigma,1000)
plt.fill_between(x1,norm.pdf(x1,mu,sigma),color='green',alpha=0.3,label='68% region (1 sigma)')

#Shade 95% for region (mu+-2)
x2=np.linspace(mu-2*sigma,mu+2*sigma,1000)
plt.fill_between(x2,norm.pdf(x2,mu,sigma),color='yellow',alpha=0.2,label='95% region (2 sigma)')

#Shade 68% for region (mu+-1)
x3=np.linspace(mu-3*sigma,mu+3*sigma,1000)
plt.fill_between(x3,norm.pdf(x3,mu,sigma),color='red',alpha=0.1,label='99.7% region (3 sigma)')

plt.title("Normal Distribution of students' score")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
