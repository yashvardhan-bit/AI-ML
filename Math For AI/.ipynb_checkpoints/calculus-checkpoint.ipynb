Math (calculus)

```python
#plot y=x**1/2
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,100,100_000)
y=np.sqrt(x)

plt.plot(x,y,label="y=x**1/2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = x^(1/2)")
plt.legend()
plt.grid()
plt.show()
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
y=np.sin(x)

plt.plot(x,y,label='y=sin(x)',c='g')

plt.axvline(0,c="b",linewidth=1.5,linestyle='--',label='x=0')
plt.axhline(0,c="b",linewidth=1.5,linestyle='--',label='y=0')

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of sin(x)')
plt.legend()
plt.show()
```
