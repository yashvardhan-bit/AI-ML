```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)  #line plot
plt.show()
```

```python
#real data

#oscar winner movies
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in Millions

#plot
plt.plot(years,oscar_revenue)
plt.title("Oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenue(in M's)")
```

```python
non_oscar_movies=["Slumdog Millionare","Avatr","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
plt.plot(years,oscar_revenue,label ="oscar movies")
plt.plot(years,non_oscar_revenue,label="non oscar movies")

plt.title("Oscar Movies Vs Non Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in M's)")
plt.legend(loc="upper left")
plt.show()
```

```python
#format string (fmt)
#fmt='[marker][line][color]'
non_oscar_movies=["Slumdog Millionare","Avatr","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
plt.plot(years,oscar_revenue,'o--g',label ="oscar movies")
plt.plot(years,non_oscar_revenue,color='red', marker='<',linestyle=':',linewidth='5',label="non oscar movies")

plt.title("Oscar Movies Vs Non Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in M's)")
plt.legend(loc="upper left")
plt.show()
```
