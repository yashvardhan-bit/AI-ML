```python
#real data

#oscar winner movies
oscar_movies=[
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232] # in Millions

#plot
plt.plot(years,oscar_revenue)
plt.title("Oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenue(in M's)")
```

```python
non_oscar_movies=["Slumdog Millionare","Avatr","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
plt.plot(years,oscar_revenue,label ="oscar movies")
plt.plot(years,non_oscar_revenue,label="non oscar movies")

plt.title("Oscar Movies Vs Non Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in M's)")
plt.legend(loc="upper left")
plt.show()
```

```python
#format string (fmt)
#fmt='[marker][line][color]'
non_oscar_movies=["Slumdog Millionare","Avatr","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,829,185,275]
plt.plot(years,oscar_revenue,'o--g',label ="oscar movies")
plt.plot(years,non_oscar_revenue,color='red', marker='<',linestyle=':',linewidth='5',label="non oscar movies")

plt.title("Oscar Movies Vs Non Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in M's)")
plt.legend(loc="upper left")
plt.show()
```

#important plot methods

```python
#Bar chart
plt.bar(years,oscar_revenue,color="r")
plt.title("Revenue in each year for oscar movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in M's $)")

#text info
for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i]+15,str(oscar_revenue[i]),ha="center")

    plt.ylim(0,max(oscar_revenue)+100)
    plt.xlim(min(years)-4,max(years)+5)
```

```python
# compare 2 data
import numpy as np
x=np.arange(len(years))
width=0.4
plt.bar(x-width/2,oscar_revenue,width,label="oscar movies")
plt.bar(x+width/2,non_oscar_revenue,width,label="non oscar movies")

plt.title("Revenue in each year for oscar movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in M's $)")
plt.legend()

plt.xticks(x,years)# replace the labels with years
```

```python
# Horizontal Chart
plt.barh(oscar_movies,oscar_revenue)
plt.xlabel("revenue")
plt.ylabel("Movie name")
```

```python
# Scatter Plot
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.scatter(age,bp)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.show()
```

```python
# customization
# color=['g'if x<135 else 'r'for x in bp]
sizes=bp
'''transparency: alpha function (range(0-1))'''
plt.scatter(age,bp,c=color,marker='>',s=bp,alpha=.4)
plt.grid()

```

```python
#cmaps
plt.scatter(age,bp,cmap="OrRd",c=(bp),s=sizes)
plt.grid()
plt.colorbar(label="BP") #scale
plt.show()
```

```python
#Annotation(text info)
# plt.annotate("text",xy=(x_point,y_point),xytext(x_text,y_text)
plt.scatter(age,bp,cmap="OrRd",c=(bp),s=sizes)
plt.grid()
plt.colorbar(label="BP") #scale

for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+1,bp[i]+1))
plt.xlim(min(age),max(age)+20)
plt.ylim(min(bp),max(bp)+5)
```

```python
# Multiple Plots
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (℃) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

plt.scatter(winter_temp,winter_humidity,label=("winter"))
plt.scatter(summer_temp,summer_humidity,label=("summer"))
plt.title("summer vs winter data")
plt.xlabel("temp")
plt.ylabel("humidity")

plt.legend()
```

```python
#Pie Plots
expenses=["Salaries","Rent","Markating","R&D","Miscelleneous"]
amount=[500,150,120,100,50]
color=['pink','r','g','b','y']
explode=[0,0,0.2,0,1]

# plt.style.available
plt.style.use("default")
plt.pie(amount,labels= expenses,autopct="%1.1f%%",colors=color
       ,wedgeprops={
           "edgecolor":"black",
           "linewidth":1,
           "linestyle":"--",
           "fill":True},
explode=explode,startangle=90,shadow=True)
plt.show()
```
