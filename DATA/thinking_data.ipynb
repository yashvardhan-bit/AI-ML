```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r")as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store_data.json")
print(data)
print(type(data))
```

```python
# Clean & structure the data
def clean_data(data):

    text_to_num = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5
    }

    cleaned_data = []
    unique_users = set()

    for user in data:

        # Clean ratings - data consistency
        raw_rating = user["rating"].strip().lower()

        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

        # Handle missing vals
        raw_age = user.get("age")

        if(raw_age == None):
            user["age"] = None

        # Deduplication
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"].strip())

        cleaned_data.append(user)

    return cleaned_data
```
