```python
#WEB SCRAPING
import requests
url="https://www.scrapethissite.com/pages/simple/"
res=requests.get(url)
```

```python
res
print(res.status_code)
```

```python
if res.status_code==200:
    print(res.headers) #meta data
```

```python
with open("scraped_data/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
#BEAUTIFULSOUP
!pip install beautifulsoup4
!pip install lxml
```

```python
from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html","r") as f:
    html_content=f.read()
soup = BeautifulSoup(html_content,"lxml")
```

```python
soup #jumbled/messy data
```

```python
soup.find("h1")# gives 1st h1 tag
```

```python
soup.find("h3")
soup.h3 #both used for first h3 tag
```

```python
all_h3=soup.find_all("h3")
type(all_h3) #a list
```

```python
for h3 in all_h3:
    print(h3)
```

```python
all_countries=[]
for h3 in all_h3:
    name=(h3.get_text(strip=True))
    # print(h3.find_parent("div")["class"])
    population=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    # print(h3.find_next("div").select_one("span.country-population").get_text(strip=True))
    print(h3.attrs)
    all_countries.append([name,population])
    print(h3.children)
```

```python
all_countries
```

```python
import pandas as pd
df=pd.DataFrame(all_countries,columns=["Naam","Jansankhya"])
```

```python
df
```

```python
df.to_csv("cleaned_data/data.csv",index=False)
```
