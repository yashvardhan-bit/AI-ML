```python
#Grouping and Aggregation

df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()

df.groupby("gender")["income"].max()

df.groupby("country")["income"].agg(["mean","min","max"])
df.groupby("country")["income"].aggregate(["mean","min","max"])#same same but different
df.groupby("country")["income"].agg(avg_salary="mean",min_salary="min",max_salary="max")#rename the aggregate values

df.groupby("country").agg({
    "income":"mean",
    "age":"mean"
})

df.groupby("country").agg({
    "income":"max",
    "age":"mean"
})

df.groupby("country").agg(
    max_salary=("income","max"),
    avg_age=("age","mean")
)
```

```python
#Melt function
'''Melt-wide to long
    
    melt(
    id_vars(keep)
    value_vars(unpivot)
    var_name(new column name)
    value_name(new column value)
    )
    '''
```

```python
df = pd.DataFrame({
"country": ["USA", "USA", "India", "India"],
"year": [2020, 2021, 2020, 2021],
"sales": [100, 120, 90, 110],
"profit": [20, 25, 18, 22],
"tax":[10,20,30,40]
})
df
```

```python
melted_df=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="matric",
    value_name="value"
)
```

```python
melted_df
```

```python
#Pivot function

#Pivot-Long to wide format
'''pivot(index,columns,values)'''
```

```python
original=melted_df.pivot(
    index=["country","year"],
    columns="matric",
    values="value"
)
```
