```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris=load_iris()
```

```python
# X=pd.DataFrame(iris.data)
# X.columns=iris.feature_names
X=iris.data
y=iris.target
```

```python
X.head()
```

```python
y
```

```python
X.shape
```

```python
# visualiza
sns.scatterplot(x=X[:,0],y=X[:,2],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
# Optional - dimensionality reduction using PCA
from sklearn.decomposition import PCA

pca=PCA(n_components=2)

pca_data=pca.fit_transform(X_scaled)
```

```python
# Elbow
wcss=[]

for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,11),y=wcss,marker='o')
```

```python
# Elbow
wcss=[]

for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
#kmeans
kmeans=KMeans(n_clusters=3)
labels=kmeans.fit_predict(pca_data)

sns.scatterplot(x=X[:,0],y=X[:,1],c=labels)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=y)
```

```python
sns.lineplot(x=range(1,11),y=wcss,marker='o')
```

```python
# KMeans

kmeans=KMeans(n_clusters=3)
labels=kmeans.fit_predict(X_scaled)

sns.scatterplot(x=X[:,0],y=X[:,2],c=labels)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,2],c=y)
```

```python
sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1],marker='x',c='b',s=500)
```

```python
# KMeans

kmeans=KMeans(n_clusters=3,random_state=10)
labels=kmeans.fit_predict(pca_data)

sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1],marker='o',c='r',s=200)
```
