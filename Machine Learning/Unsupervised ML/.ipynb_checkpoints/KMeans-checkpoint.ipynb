```python
import seaborn as sns
from sklearn.datasets import make_blobs # to produce cluster data
```

```python
X,y=make_blobs(
    n_samples=1000,
    n_features=2,
    centers=4, # 4 true cluster
    random_state=42
)
```

```python
y
```

```python
# Visualize 
sns.scatterplot(x=X[:,0],y=X[:,1]) # 0th and 1st feature 
```

```python
# Scaling is important
```

```python
# K-Means Clustering
from sklearn.cluster import KMeans
```

```python
K=4
kmeans=KMeans(
        n_clusters=K, # default=8
        random_state=42
)
```

```python
labels =kmeans.fit_predict(X)
#label = cluster number
```

```python
labels
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=labels)
```

### Choose our K vals - elblow; silhouette score

```python
# Elbow Method

wcss=[]
for k in range(1,21):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,21),y=wcss,marker='*',markersize=10)
```

```python
# Kneed Module 
!pip install kneed
```

```python
from kneed import KneeLocator
```

```python
knee=KneeLocator(range(1,21),wcss,curve="convex",direction="decreasing")
```

```python
print("optimal K :",knee.elbow)
```

### Silhouette Score

```python
from sklearn.metrics import silhouette_score
```

```python
ss=[]
for k in range(2,21): # always start with 2
    kmeans=KMeans(n_clusters=k)
    labels=kmeans.fit_predict(X)
    score=silhouette_score(X,labels)
    ss.append(score)
```

```python
# plot -K and ss

sns.lineplot(x=range(2,21),y=ss,marker='o')
```
